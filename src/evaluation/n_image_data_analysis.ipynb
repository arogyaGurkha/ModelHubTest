```python
import os
# Import all the csvs



def list_directories(directory):
    subdirectories = sorted(
        [d for d in os.listdir(directory) if os.path.isdir(os.path.join(directory, d))],
        reverse=True,
    )
    return subdirectories


root_directory = "/workspaces/ModelHubTest/src/data/experiments/n_image_inference/May-21/Extreme Experiment"
subdirectories = list_directories(root_directory)

# n_500_df = n_500_df.drop(['Unnamed: 0'], axis=1)
# n_500_df = n_500_df.sort_values(by=['predicted_accuracy'], ascending=False)

csvs = []
for sub in subdirectories:
    csv = (
        pd.read_csv(f"{root_directory}/{sub}/test_results.csv")
        .drop(["Unnamed: 0"], axis=1)
        .sort_values(by=["predicted_accuracy"], ascending=False)
    )
    csvs.append(csv)

csvs
```

```python
# Combine all data
extreme_analysis_df = ground_truth_df

for df, sub in zip(csvs, subdirectories):
    extreme_analysis_df = pd.merge(
        extreme_analysis_df,
        df[["model", "predicted_accuracy"]].rename(
            columns={"predicted_accuracy": f"{sub}_predicted_accuracy"}
        ),
        on="model",
        how="left",
    )
    # extreme_analysis_df[f"{sub}_predicted_accuracy"].fillna(extreme_analysis_df[f"{sub}_predicted_accuracy"].mean(), inplace=True)

extreme_analysis_df = extreme_analysis_df.dropna()

extreme_analysis_df
```

```python
# Calculate Error

mae = []
mse = []
r2 = []
for sub in subdirectories:
    mae.append(mean_absolute_error(extreme_analysis_df['groundt_accuracy'], extreme_analysis_df[f'{sub}_predicted_accuracy']))
    mse.append(mean_squared_error(extreme_analysis_df['groundt_accuracy'], extreme_analysis_df[f'{sub}_predicted_accuracy']))
    r2.append(r2_score(extreme_analysis_df['groundt_accuracy'], extreme_analysis_df[f'{sub}_predicted_accuracy']))


data = {
    'n': [i for i in range(50, 0, -1)],
    'MAE': mae,
    'MSE': mse, 
    'r2': r2,
}

df = pd.DataFrame(data)
df
```

```python
# Plot
plt.figure(figsize=(12, 8))

# Plot MAE
plt.plot(df['n'], df['MAE'], marker='o', label='MAE')
# Plot MSE
plt.plot(df['n'], df['MSE'], marker='s', label='MSE')
# Plot r2
plt.plot(df['n'], df['r2'], marker='^', label='R2')

plt.xlabel('Sample Count (n)')
plt.ylabel('Metric Values')
plt.title('Metrics Trend as Sample Count Decreases')
plt.grid(True)
plt.xticks(df['n'])  # Ensure all x values are shown
plt.gca().invert_xaxis()
plt.legend()

plt.show()
```

```python
# r2 vs mse, mae

fig, ax1 = plt.subplots(figsize=(12, 8))

# Plot MAE and MSE on the primary y-axis
color = 'tab:blue'
ax1.set_xlabel('Sample Count (n)')
ax1.set_ylabel('MAE / MSE', color=color)
ax1.plot(df['n'], df['MAE'], marker='o', label='MAE', color='tab:blue')
ax1.plot(df['n'], df['MSE'], marker='s', label='MSE', color='tab:cyan')
ax1.tick_params(axis='y', labelcolor=color)
ax1.invert_xaxis()
ax1.grid(True)
ax1.legend(loc='upper left')

# Create a second y-axis for r2
ax2 = ax1.twinx()  
color = 'tab:red'
ax2.set_ylabel('R2', color=color)
ax2.plot(df['n'], df['r2'], marker='^', label='R2', color='tab:red')
ax2.tick_params(axis='y', labelcolor=color)
ax2.legend(loc='upper right')

plt.title('Trend of MAE, MSE, and R2 as Sample Count Decreases')
plt.xticks(df['n'])  # Ensure all x values are shown

fig.tight_layout()  # Adjust layout to accommodate both y-axes

plt.show()
```

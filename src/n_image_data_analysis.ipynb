```python
analysis_df = ground_truth_df
for df, label in zip([n_3618_df, n_3000_df, n_2500_df, n_2000_df, n_1500_df, n_1000_df, n_500_df], ['n_3618_df', 'n_3000_df', 'n_2500_df', 'n_2000_df', 'n_1500_df', 'n_1000_df', 'n_500_df']):
    analysis_df = pd.merge(analysis_df, df[['model', 'predicted_precision']].rename(
        columns={'predicted_precision': f'{label}_predicted_precision'}
    ), on='model', how='left')
analysis_df = analysis_df.sort_values(['n_3618_df_predicted_precision'], ascending=False)
analysis_df
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np
```

```python

for col in [c for c in analysis_df.columns if 'predicted_precision' in c]:
    # Calculate absolute error for each entry
    analysis_df[f'abs_error_{col}'] = (analysis_df['pred_precision'] - analysis_df[col]).abs()

    # Calculate squared error for each entry
    analysis_df[f'squared_error_{col}'] = (analysis_df['pred_precision'] - analysis_df[col]) ** 2

analysis_df.to_csv("/workspaces/ModelHubTest/src/data/experiments/n_image_inference/prediction_analysis.csv")
```
